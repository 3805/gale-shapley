```python
WOMEN = ['AMY', 'BERTHA', 'CLARE', 'DIANE', 'ERIKA']
MEN = ['VICTOR', 'WYATT', 'XAVIER', 'YANCEY', 'ZEUS']
p = {}

class Person: 
    def __init__(self, name, order, so=None):
        self.name = name
        self.order = order
        self.so = so

def create_var(name, order):
    p[name] = Person(name, order)

[create_var(name, order) for order, name in enumerate(WOMEN)]
[create_var(name, order) for order, name in enumerate(MEN)]

mpl = [[p['BERTHA'], p['AMY'], p['DIANE'], p['ERIKA'], p['CLARE']],
       [p['DIANE'], p['BERTHA'], p['AMY'], p['CLARE'], p['ERIKA']],
       [p['BERTHA'], p['ERIKA'], p['CLARE'], p['DIANE'], p['AMY']],
       [p['AMY'], p['DIANE'], p['CLARE'], p['BERTHA'], p['ERIKA']],
       [p['BERTHA'], p['DIANE'], p['AMY'], p['ERIKA'], p['CLARE']]]

wpl = [[p['ZEUS'], p['VICTOR'], p['WYATT'], p['YANCEY'], p['XAVIER']],
       [p['XAVIER'], p['WYATT'], p['YANCEY'], p['VICTOR'], p['ZEUS']],
       [p['WYATT'], p['XAVIER'], p['YANCEY'], p['ZEUS'], p['VICTOR']],
       [p['VICTOR'], p['ZEUS'], p['YANCEY'], p['XAVIER'], p['WYATT']],
       [p['YANCEY'], p['WYATT'], p['ZEUS'], p['XAVIER'], p['VICTOR']]]

def reply(keep_pretendent, person, o):
    nm_name = person['nm']
    pm_name = person['pm']
    w_name = person['w']
    
    if keep_pretendent:
        p[w_name].so = nm_name
        
        pm_id = MEN.index(pm_name)
        pm_pl = [woman.name for woman in mpl[pm_id]]
        
        w_priority = pm_pl.index(w_name)
        
        if w_priority < 5:
            return choose_by_priority(w_priority+1, pm_id)
        
        if o['man'] < 5:
            return choose_by_priority(o['priority'], o['man']+1)
        
    if not keep_pretendent:
        if o['priority'] < 5:
            return choose_by_priority(o['priority']+1, o['man'])
        
        
def propose(m_name, w_name, o):
    w = p[w_name]
    m = p[m_name]
    w_status = w.so

    if w_status is None:
        w.so = m_name
        choose_by_priority(0, o['man']+1)
    
    if w_status is not None:
        husband_name = w.so
        
        if m_name == husband_name:
            return choose_by_priority(0, o['man']+1)
            
        w_pl = [g.name for g in wpl[w.order]]
        
        nm_priority = w_pl.index(m_name)
        pm_priority = w_pl.index(husband_name)
        
        reply(True if pm_priority > nm_priority else False, 
              {'w': w_name, 'nm': m_name, 'pm': husband_name},
              o)
        
def choose_by_priority(priority, man):
    if man < 5:
        w_name = mpl[man][priority].name
        m_name = MEN[man]
        propose(m_name, w_name, {"priority": priority, "man": man})

choose_by_priority(0, 0)
```
